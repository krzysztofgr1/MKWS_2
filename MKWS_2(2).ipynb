```python
import cantera as ct
import SDToolbox as sd
import matplotlib.pyplot as plt
import numpy as np

P0=1
T0=300
mech = 'gri30.xml'

pressure_array = np.linspace(0.5,8,20)
temperature_array = np.linspace(300,760,20)

#variable pressure

pressure=[]
temperature=[]
speed=[]
C=[]
i=0

for P0 in pressure_array: 

    P0 *= ct.one_atm
    x = 0.8124 # max
    a=4.76*x/(1-x)
    q ='O2:1.,N2:3.76,H2:' + str(a)
    
    
    [cj_speed,R2] = sd.CJspeed(P0, T0, q, mech, 0)
    U1 = cj_speed 

    gas = sd.PostShock_eq(cj_speed, P0, T0, q, mech)
    Ps = gas.P/ct.one_atm

    C.append(x*100)
    
    
    speed.append(cj_speed)
    i=i+1

plt.plot(pressure_array,speed)
plt.xlabel('Initial Pressure [atm]')
plt.ylabel('C-J Speed [m/s]')
plt.show()


#variable concentration

pressure=[]
temperature=[]
speed=[]
C=[]
i = 0
P0 = ct.one_atm

while i < 15:
    nH2 = 0.05 + float(i)/20
    nO2 = 0.5*(1-nH2) /4.76
    nN2 = 0.5*(1-nH2) *3.76/4.76
    X  =  'H2:'+str(nH2)+ ' ' + 'O2:'+str(nO2) + ' ' + 'N2:'+str(nN2)
    C.append(nH2)
    [cj_speed,R2] = sd.CJspeed(P0, T0, X, mech, 0);
    gas = sd.PostShock_eq(cj_speed, P0, T0, X, mech)
    
    speed.append(cj_speed)
    
    i = i + 1
    
plt.plot(C,speed)
plt.xlabel('Concentration [-]')
plt.ylabel('C-J Speed [m/s]')
plt.show()


#variable temperature

pressure=[]
temperature=[]
speed=[]
C=[]
i=0
for T0 in temperature_array:
    P0 = ct.one_atm
    x = 0.1065 # value of max
    a=4.76*x/(1-x)
    q ='O2:1.,N2:3.76,H2:' + str(a)
    
    
    [cj_speed,R2] = sd.CJspeed(P0, T0, q, mech, 0)
    U1 = cj_speed  #shock speed

    gas = sd.PostShock_eq(cj_speed, P0, T0, q, mech)
    Ps = gas.P/ct.one_atm

    C.append(x*100)
   
    speed.append(cj_speed)
    i=i+1

#plots    
    
plt.plot(temperature_array,speed)
plt.xlabel('Initial temperature [K]')
plt.ylabel('C-J Speed [m/s]')
plt.show()
```
